```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="gemma2:2b")
)
```

```python
r1 = embeddings.embed_documents(
    [
        "The cat sat on the mat",
        "The dog chased the cat",
        "The cat was happy with the dog",
    ]
)
```
